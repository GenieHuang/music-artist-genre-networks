```python
top10_genres = genre_df.nlargest(10, 'popularity')
top10_genres
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# The musical features of top 10 genres
features = ['acousticness', 'danceability', 'energy', 'liveness']

melted_data = pd.melt(top10_genres, 
                      id_vars=['genres'], 
                      value_vars=features,
                      var_name='feature', 
                      value_name='value')

plt.figure(figsize=(12, 6))
ax = sns.barplot(x='genres', y='value', hue='feature', data=melted_data)

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3)

ax.set_title('Top 10 Genres by Popularity')
ax.set_ylabel('Values')
ax.set_ylim(0, 1.2)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
